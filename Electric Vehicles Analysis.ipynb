        1. List the top 3 and bottom 3 makers for the fiscal years 2023 and 2024 in terms of the number of 2-wheelers sold
    </h2>
</div>


```python

df_maker_sales['date'] = pd.to_datetime(df_maker_sales['date'])


def get_fiscal_year(date):
    if date.month >= 7:
        return date.year + 1
    else:
        return date.year

df_maker_sales['fiscal_year'] = df_maker_sales['date'].apply(get_fiscal_year)

```

```python

df_maker_sales['fiscal_year'] = df_maker_sales['date'].apply(get_fiscal_year)


result = df_maker_sales.groupby(['maker', 'fiscal_year']).agg(
    electric_vehicles_sold=('electric_vehicles_sold', 'sum')
).reset_index()


top_2023 = result[result['fiscal_year'] == 2023].sort_values(by='electric_vehicles_sold',ascending=False). head(3)
bottom_2023 = result[result['fiscal_year'] == 2023].sort_values(by='electric_vehicles_sold',ascending=False).tail(3)
top_2024 = result[result['fiscal_year'] == 2024].sort_values(by='electric_vehicles_sold', ascending=False).head(3)
bottom_2024 = result[result['fiscal_year'] == 2024].sort_values(by='electric_vehicles_sold',ascending=False).tail(3)


fig, axes = plt.subplots(2, 2, figsize=(18, 8))


def add_bar_labels(ax):
    for container in ax.containers:
        ax.bar_label(container, label_type='edge')

# Top 3 makers for 2023
axes[0, 0].bar(top_2023['maker'], top_2023['electric_vehicles_sold'], color='blue')
axes[0, 0].set_title('Top 3 Makers for 2023')
axes[0, 0].set_xlabel('Maker')
axes[0, 0].set_ylabel('Electric Vehicles Sold')
add_bar_labels(axes[0, 0])

# Bottom 3 makers for 2023
axes[0, 1].bar(bottom_2023['maker'], bottom_2023['electric_vehicles_sold'], color='red')
axes[0, 1].set_title('Bottom 3 Makers for 2023')
axes[0, 1].set_xlabel('Maker')
axes[0, 1].set_ylabel('Electric Vehicles Sold')
add_bar_labels(axes[0, 1])

# Top 3 makers for 2024
axes[1, 0].bar(top_2024['maker'], top_2024['electric_vehicles_sold'], color='green')
axes[1, 0].set_title('Top 3 Makers for 2024')
axes[1, 0].set_xlabel('Maker')
axes[1, 0].set_ylabel('Electric Vehicles Sold')
add_bar_labels(axes[1, 0])

# Bottom 3 makers for 2024
axes[1, 1].bar(bottom_2024['maker'], bottom_2024['electric_vehicles_sold'], color='orange')
axes[1, 1].set_title('Bottom 3 Makers for 2024')
axes[1, 1].set_xlabel('Maker')
axes[1, 1].set_ylabel('Electric Vehicles Sold')
add_bar_labels(axes[1, 1])

plt.tight_layout()
plt.show()
```

```python

df_state_sales ['date'] = pd.to_datetime(df_state_sales ['date'])


def get_fiscal_year(date):
    if date.month >= 7:
        return date.year + 1
    else:
        return date.year

df_state_sales ['fiscal_year'] = df_state_sales ['date'].apply(get_fiscal_year)


```

```python
df_state_sales['date'] = pd.to_datetime(df_state_sales['date'])


df_2024 = df_state_sales[df_state_sales['date'].dt.year == 2024]


df_filtered = df_2024[df_2024['vehicle_category'].isin(['2-Wheelers', '4-Wheelers'])]

df_grouped = df_filtered.groupby(['state', 'vehicle_category']).agg(
    electric_vehicles_sold=('electric_vehicles_sold', 'sum'),
    total_vehicles_sold=('total_vehicles_sold', 'sum')
).reset_index()

df_grouped['penetration_rate'] = df_grouped['electric_vehicles_sold'] / df_grouped['total_vehicles_sold']


df_2w = df_grouped[df_grouped['vehicle_category'] == '2-Wheelers']
df_4w = df_grouped[df_grouped['vehicle_category'] == '4-Wheelers']


top_5_states_2w = df_2w.sort_values(by='penetration_rate', ascending=False).head(5)


top_5_states_4w = df_4w.sort_values(by='penetration_rate', ascending=False).head(5)


def add_bar_labels(ax):
    for container in ax.containers:
        ax.bar_label(container, label_type='edge')


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 5), sharey=True)


top_5_states_2w.plot(kind='bar', x='state', y='penetration_rate', ax=ax1, color='skyblue', legend=False)
ax1.set_title('Top 5 States for 2-Wheeler EV Penetration Rate in FY 2024')
ax1.set_xlabel('State')
ax1.set_ylabel('Penetration Rate')
add_bar_labels(ax1)
ax1.tick_params(axis='x', rotation=45)  # Rotate x-axis labels by 45 degrees


top_5_states_4w.plot(kind='bar', x='state', y='penetration_rate', ax=ax2, color='lightgreen', legend=False)
ax2.set_title('Top 5 States for 4-Wheeler EV Penetration Rate in FY 2024')
ax2.set_xlabel('State')
ax2.set_ylabel('Penetration Rate')
add_bar_labels(ax2)
ax2.tick_params(axis='x', rotation=45)  # Rotate x-axis labels by 45 degrees

plt.tight_layout()
plt.show()
```

```python

df_state_sales['date'] = pd.to_datetime(df_state_sales['date'])


df_filtered = df_state_sales[df_state_sales['date'].dt.year.isin([2022, 2023, 2024])]


df_filtered = df_filtered[df_filtered['vehicle_category'].isin(['2-Wheelers', '4-Wheelers'])]


df_grouped = df_filtered.groupby(['state', df_filtered['date'].dt.year]).agg(
    electric_vehicles_sold=('electric_vehicles_sold', 'sum'),
    total_vehicles_sold=('total_vehicles_sold', 'sum')
).reset_index()


df_grouped['penetration_rate'] = df_grouped['electric_vehicles_sold'] / df_grouped['total_vehicles_sold']


df_pivot = df_grouped.pivot(index='state', columns='date', values='penetration_rate')


df_pivot.columns = [str(col) for col in df_pivot.columns]


df_pivot['decline_2023_2022'] = df_pivot['2022'] - df_pivot['2023']
df_pivot['decline_2024_2023'] = df_pivot['2023'] - df_pivot['2024']

decline_states = df_pivot[(df_pivot['decline_2023_2022'] < 0) | (df_pivot['decline_2024_2023'] < 0)]

print("States with Negative Penetration Rate Changes:")
print(decline_states[['decline_2023_2022', 'decline_2024_2023']])
```

```python

df_maker_sales['date'] = pd.to_datetime(df_maker_sales['date'])


df_filtered = df_maker_sales[(df_maker_sales['vehicle_category'] == '4-Wheelers') & 
                             (df_maker_sales['date'].dt.year.isin([2022, 2024]))]


df_grouped = df_filtered.groupby(['maker', df_filtered['date'].dt.year]).agg(
    electric_vehicles_sold=('electric_vehicles_sold', 'sum')
).reset_index()


df_pivot = df_grouped.pivot(index='maker', columns='date', values='electric_vehicles_sold')


def calculate_cagr(start_value, end_value, periods):
    if start_value == 0:
        return float('inf')  # Handle division by zero
    return (end_value / start_value) ** (1 / periods) - 1

df_pivot['CAGR'] = df_pivot.apply(lambda row: calculate_cagr(row[2022], row[2024], 2), axis=1)


df_pivot['Total_Sales'] = df_pivot[2022] + df_pivot[2024]
top_5_makers = df_pivot.nlargest(5, 'Total_Sales')
print(top_5_makers)
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(top_5_makers.index, top_5_makers['CAGR'], marker='o', linestyle='-', color='blue')

# Add labels to each point
for i, txt in enumerate(top_5_makers['CAGR']):
    ax.annotate(f'{txt:.1%}', (top_5_makers.index[i], top_5_makers['CAGR'].iloc[i]),
                textcoords="offset points", xytext=(0,12), ha='center')

ax.set_title('CAGR in 4-Wheeler Units for Top 5 Makers (2022 to 2024)')
ax.set_xlabel('Maker')
ax.set_ylabel('CAGR')
ax.set_xticks(range(len(top_5_makers.index)))
ax.set_xticklabels(top_5_makers.index, rotation=45)

plt.tight_layout()
plt.show()
```

```python

df_state_sales['date'] = pd.to_datetime(df_state_sales['date'])


df_filtered = df_state_sales[df_state_sales['date'].dt.year.isin([2022, 2023, 2024])]


df_filtered['month'] = df_filtered['date'].dt.to_period('M')


df_monthly_sales = df_filtered.groupby('month').agg(
    total_electric_vehicles_sold=('electric_vehicles_sold', 'sum')
).reset_index()


df_monthly_sales['month'] = df_monthly_sales['month'].astype(str)


peak_month = df_monthly_sales.loc[df_monthly_sales['total_electric_vehicles_sold'].idxmax()]
low_month = df_monthly_sales.loc[df_monthly_sales['total_electric_vehicles_sold'].idxmin()]

print(f"Peak Season Month: {peak_month['month']} with Sales: {peak_month['total_electric_vehicles_sold']}")
print(f"Low Season Month: {low_month['month']} with Sales: {low_month['total_electric_vehicles_sold']}")


fig, ax = plt.subplots(figsize=(17, 7))


ax.plot(df_monthly_sales['month'], df_monthly_sales['total_electric_vehicles_sold'], marker='o', color='b', linestyle='-', linewidth=2, markersize=8)


ax.scatter(peak_month['month'], peak_month['total_electric_vehicles_sold'], color='r', s=100, zorder=5, label='Peak Month')
ax.scatter(low_month['month'], low_month['total_electric_vehicles_sold'], color='g', s=100, zorder=5, label='Low Month')


for i, row in df_monthly_sales.iterrows():
    ax.annotate(f'{row["total_electric_vehicles_sold"]}', 
                (row['month'], row['total_electric_vehicles_sold']),
                textcoords="offset points", xytext=(0,10), ha='center')


ax.annotate(f'{peak_month["total_electric_vehicles_sold"]}', 
            (peak_month['month'], peak_month['total_electric_vehicles_sold']),
            textcoords="offset points", xytext=(0,10), ha='center', color='r')
ax.annotate(f'{low_month["total_electric_vehicles_sold"]}', 
            (low_month['month'], low_month['total_electric_vehicles_sold']),
            textcoords="offset points", xytext=(0,10), ha='center', color='g')


ax.set_xlabel('Month')
ax.set_ylabel('Total Electric Vehicles Sold')
ax.set_title('Monthly EV Sales from 2022 to 2024')


plt.xticks(rotation=45)
ax.grid(True)


ax.legend()

plt.tight_layout()
plt.show()
```

        9. What is the projected number of EV sales (including 2-wheelers and 4-wheelers) for the top 10 states by penetration rate in 2030, based on the compounded annual growth rate (CAGR) from previous years?
    </h2>
</div>


```python

df_state_sales['date'] = pd.to_datetime(df_state_sales['date'])


df_filtered = df_state_sales[df_state_sales['date'].dt.year.isin([2021, 2022, 2023, 2024])]


df_grouped = df_filtered.groupby(['state', df_filtered['date'].dt.year]).agg(
    electric_vehicles_sold=('electric_vehicles_sold', 'sum')
).reset_index()


df_pivot = df_grouped.pivot(index='state', columns='date', values='electric_vehicles_sold')


def calculate_cagr(start_value, end_value, periods):
    if start_value == 0:
        return np.nan  # Handle division by zero or undefined CAGR
    return (end_value / start_value) ** (1 / periods) - 1


df_pivot['CAGR_2021_2024'] = df_pivot.apply(lambda row: calculate_cagr(row[2021], row[2024], 3), axis=1)


top_10_states = df_pivot.nlargest(10, 'CAGR_2021_2024').reset_index()


latest_year = 2024
top_10_states['projected_sales_2030'] = top_10_states.apply(lambda row: row[latest_year] * ((1 + row['CAGR_2021_2024']) ** (2030 - latest_year)), axis=1)


top_10_states = top_10_states.sort_values(by='projected_sales_2030', ascending=False)

print("Projected EV Sales for Top 10 States in 2030:")
print(top_10_states[['state', 'projected_sales_2030']])


fig, ax = plt.subplots(figsize=(16, 5))


ax.plot(top_10_states['state'], top_10_states['projected_sales_2030'], marker='o', linestyle='-', color='b', label='Projected Sales 2030')

for i, value in enumerate(top_10_states['projected_sales_2030']):
    ax.text(i, value, f'{value:,.0f}', ha='center', va='bottom', fontsize=10)


ax.set_xlabel('State')
ax.set_ylabel('Projected EV Sales (2030)')
ax.set_title('Projected EV Sales for Top 10 States by Penetration Rate in 2030')


plt.xticks(rotation=45)


ax.grid(True)


ax.legend()

plt.tight_layout()
plt.show()
```

```python
df_2_wheelers_FY = df_state_sales[df_state_sales['vehicle_category'] == '2-Wheelers'].groupby(['fiscal_year'])['electric_vehicles_sold'].sum().reset_index()
df_2_wheelers_FY
```

```python
df_4_wheelers_FY = df_state_sales[df_state_sales['vehicle_category'] == '4-Wheelers'].groupby(['fiscal_year'])['electric_vehicles_sold'].sum().reset_index()
df_4_wheelers_FY
```

```python
df_2_wheelers_FY = df_state_sales[df_state_sales['vehicle_category'] == '2-Wheelers'].groupby('fiscal_year')['electric_vehicles_sold'].sum()
df_4_wheelers_FY = df_state_sales[df_state_sales['vehicle_category'] == '4-Wheelers'].groupby('fiscal_year')['electric_vehicles_sold'].sum()

price_2w = 85000
price_4w = 1500000

# Function to calculate revenue growth rates
def calculate_growth_rate(start_year, end_year, df, price):
    revenue_start = df.get(start_year, 0) * price
    revenue_end = df.get(end_year, 0) * price
    if revenue_start == 0:
        return float('inf')
    return round(((revenue_end - revenue_start) / revenue_start) * 100, 2)


growth_rate_2w_22_24 = calculate_growth_rate(2022, 2024, df_2_wheelers_FY, price_2w)
growth_rate_2w_23_24 = calculate_growth_rate(2023, 2024, df_2_wheelers_FY, price_2w)
growth_rate_4w_22_24 = calculate_growth_rate(2022, 2024, df_4_wheelers_FY, price_4w)
growth_rate_4w_23_24 = calculate_growth_rate(2023, 2024, df_4_wheelers_FY, price_4w)


print(f"Growth rate of revenue from 2-wheelers (FY2022 to FY2024): {growth_rate_2w_22_24}%")
print(f"Growth rate of revenue from 2-wheelers (FY2023 to FY2024): {growth_rate_2w_23_24}%")
print(f"Growth rate of revenue from 4-wheelers (FY2022 to FY2024): {growth_rate_4w_22_24}%")
print(f"Growth rate of revenue from 4-wheelers (FY2023 to FY2024): {growth_rate_4w_23_24}%")
```

```python

years = ['FY2022 to FY2024', 'FY2023 to FY2024']
x = range(len(years))

fig, ax = plt.subplots(figsize=(8, 4))

# Scatter plot with lines
ax.plot(x, [growth_rate_2w_22_24, growth_rate_2w_23_24], marker='o', color='b', linestyle='-', linewidth=2, markersize=8, label='2-Wheelers')
ax.plot(x, [growth_rate_4w_22_24, growth_rate_4w_23_24], marker='o', color='g', linestyle='-', linewidth=2, markersize=8, label='4-Wheelers')


for i in x:
    ax.text(i, growth_rate_2w_22_24 if i == 0 else growth_rate_2w_23_24, f'{growth_rate_2w_22_24 if i == 0 else growth_rate_2w_23_24}%', ha='center', color='black')
    ax.text(i, growth_rate_4w_22_24 if i == 0 else growth_rate_4w_23_24, f'{growth_rate_4w_22_24 if i == 0 else growth_rate_4w_23_24}%', ha='center', color='black')


ax.set_xticks(x)
ax.set_xticklabels(years)
ax.set_xlabel('Period')
ax.set_ylabel('Growth Rate (%)')
ax.set_title('Revenue Growth Rate of 2-Wheelers and 4-Wheelers (Scatter with Lines)')
ax.legend()

plt.tight_layout()
plt.show()
```
